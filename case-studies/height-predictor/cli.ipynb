```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import preprocessing

data = pd.read_csv('weight-height.csv')
```

```python
data.head()
sns.scatterplot(data=data, x='Weight', y='Height', hue='Gender')
data.shape
```

```python
data_male = data[data['Gender']=='Male']
data_male.shape
```

```python
data_male.head()
```

```python
data_male2 = data_male.dropna()
data_male2.shape
```

```python
weights = data_male['Weight'].values
weights.shape
```

```python
heights = data_male['Height'].values.reshape((-1,1))
heights.shape
```

```python
model = LinearRegression().fit(heights,weights)
```

```python
print(model.coef_)
print(model.intercept_)
def line_eq(x):
    return model.coef_[0]*x+model.intercept_

x = np.linspace(50,80,100)
y = line_eq(x)
```

```python
plt.scatter(heights,weights)
plt.plot(x,y,color="limegreen")
```

```python
model.score(heights,weights)
```
